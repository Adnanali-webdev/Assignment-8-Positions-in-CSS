Q2) Problem Statement

   Explain the difference between Absolute and Relative positioning.

Ans)  The difference between Absolute and Relative are:

relative

1)  With relative positioning, an element can be moved up,
down, left, or right from its normal position.

2) It will not break the normal document flow to position
the element on the page.

3) The element will leave the space at its original position
& other content will not be adjusted to fit into any gap
left by the element.

absolute

1) With absolute positioning, an element is positioned relative
to its nearest positioned ancestor.

2) It will break the normal document flow to position
the element on the page.

3) The element will not leave any space in its original
position.

Q5) Explain the z-index, with a code example.

Ans) The z-index CSS property arranges the z-position of a positioned element, as well as its descendants.

while placing our element on a webpage, there will be case of overlap.
z-index specifies a stack order of our element.

The z-index property takes an integer value, which represents the element's position on the z-axis. Elements with higher z-index values are placed on top of elements with lower z-index values.

By default, elements have a z-index value of 0

```python
from IPython.display import display, HTML

html_code = """
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Task-5</title>
    <style>
        *{
            margin: 0;
            padding: 0;
        }
        .boxes{
            position:relative;
        }
        .box{
            width:200px;
            height: 200px;
            border:2px solid black;
            text-align: center;
        }
        .box1{
            background-color: red;
            position: absolute;
            left:300px;
            top:50px;
        }
        .box2{
            background-color: green;
            position: absolute;
            left:450px;
            z-index:1;
        }
        .box3{
            background-color: yellow;
            position: absolute;
            left:300px;
            top:150px;
            z-index: 2;
        }
    </style>
</head>
<body>
    <div class="boxes">
        <div class="box1 box">1</div>
        <div class="box2 box">2</div>
        <div class="box3 box">3</div>
    </div>

</body>
</html>
"""

# Display the HTML content
display(HTML(html_code))
```
